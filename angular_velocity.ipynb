```python
import pandas as pd #for import csv
```

```python
angular_data = pd.read_csv("1103.csv") #before 'upload files'
```

```python
import matplotlib.pyplot as plt
```

```python
x = angular_data['data1']
y_angX = angular_data['data2']
y_angY = angular_data['data3']
y_angZ = angular_data['data4']
```

```python
fig, ax1 = plt.subplots()
p1 = ax1.plot(x,y_angX,label='pop',marker="o",color="red")
ax1.set_title("angular data")
ax1.set_xlabel('time')
ax1.set_ylabel('red rad/s')
ax2 = ax1.twinx()  # 2つのプロットを関連付ける
ax3 = ax1.twinx()  # 2つのプロットを関連付ける
rspine = ax3.spines['right'] # 3軸対応
rspine.set_position(('axes', 1.2)) # y軸の位置をずらす
#ax3.get_yaxis().get_major_formatter().set_useOffset(False) # 気圧の数字を省略形から本来の形へ 
p2 = ax2.plot(x,y_angY,label='pop',marker="o",color="green")
ax2.set_ylabel('green')
p3 = ax3.plot(x,y_angZ,label='pop',marker="o",color="blue")
ax3.set_ylabel('blue')
ax1.legend((p1[0], p2[0],p3[0]), ("angX", "angY","angZ"), loc=2)
```
